## Name

```python
title = df_train.name.apply(func = lambda i: i.split(',')[1].split('.')[0].strip())
```

## age

```python
df_train[df_train.age.isna()]
```

```python
media_age_title = df_train.groupby(by = 'title', sort = False)['age'].mean().round()
```

```python
idade_na_index = df_train.age[df_train.age.isna()].index

titulo_na = df_train.title[idade_na]

titulo_na.apply(func = lambda x: media_age_title[x])


for i in titulo_na.apply(func = lambda x: media_age_title[x]).index:
    df_train.loc[ i ,'age'] = titulo_na.apply(func = lambda x: media_age_title[x])[i]
    
df_train
```
